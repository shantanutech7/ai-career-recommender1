```python
import pandas as pd
df = pd.read_csv("career_data.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder

# Create encoders
skill_enc = LabelEncoder()
interest_enc = LabelEncoder()
style_enc = LabelEncoder()
career_enc = LabelEncoder()

# Apply encoders to each column
df["skills_encoded"] = skill_enc.fit_transform(df["skills"])
df["interests_encoded"] = interest_enc.fit_transform(df["interests"])
df["work_style_encoded"] = style_enc.fit_transform(df["work_style"])
df["career_encoded"] = career_enc.fit_transform(df["career_path"])

df
```

```python
from sklearn.tree import DecisionTreeClassifier

# Features (X) and target (y)
X = df[["skills_encoded", "interests_encoded", "work_style_encoded"]]
y = df["career_encoded"]

# Train model on full dataset (because dataset is small)
model = DecisionTreeClassifier()
model.fit(X, y)

print("Model trained successfully!")
```

```python
import joblib

# Save model
joblib.dump(model, "src/career_model.pkl")

# Save encoders
joblib.dump(skill_enc, "src/skill_encoder.pkl")
joblib.dump(interest_enc, "src/interest_encoder.pkl")
joblib.dump(style_enc, "src/style_encoder.pkl")
joblib.dump(career_enc, "src/career_encoder.pkl")

print("Model and encoders saved successfully!")
```

```python
def recommend_career(skills, interests, work_style):
    s = skill_enc.transform([skills])[0]
    i = interest_enc.transform([interests])[0]
    w = style_enc.transform([work_style])[0]
    
    pred = model.predict([[s, i, w]])[0]
    return career_enc.inverse_transform([pred])[0]

# Quick test
recommend_career("python", "machine learning", "independent")
```
